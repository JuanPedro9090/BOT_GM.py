## Agrupamento dos dados através da "Grupo de Mercadoria" e contando a quantidade

```python
df_quant2 = ProdutoGeral['Grupo_merc'].value_counts().to_frame()
df_quant2.columns=['Quantidade']
```

```python
df_quant2
```

```python
df_final5 = ProdutoGeral.groupby('Grupo_merc').sum()
df_final5
```
